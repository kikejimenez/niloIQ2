# Datetime
> Rename and format column `fecha` as `date`

```python
#| default_exp etl.international.to_datetime
```

```python
#| export
import os
import pathlib
import pandas as pd
import unidecode
```

```python
#| export

from etl import tables
```

```python
#| export
old_version = '01'
new_version = '02'

for table in tables.international_list:

    df = pd.read_csv(
        table.version(old_version),
        low_memory=False,
    )

    df['date'] = pd.to_datetime(
        df['fecha'],
        infer_datetime_format=True,
    )
    df.drop(columns=['fecha'],inplace=True)
    df.set_index('date', inplace=True)


    df.to_csv(table.version(new_version))
```
