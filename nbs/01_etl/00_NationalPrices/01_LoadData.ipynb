```python
#| export
from sqlalchemy import create_engine
from niloIQ2.etl import tables

import pandas as pd
import os
import pathlib
from dotenv import load_dotenv
```

```python
tables.national_list
```

```python
#| export

load_dotenv()

USER = os.getenv('USER')
PASS = os.getenv('PASS')
HOST = os.getenv('IP')
PORT = os.getenv('PORT')
DB = os.getenv('DB')

sqlEngine = create_engine(
    f'mysql+pymysql://{USER}:{PASS}@{HOST}/{DB}', pool_recycle=3600
)
dbConnection = sqlEngine.connect()


new_version ='01'

for table in tables.national_list:
    df = pd.read_sql(f"select * from {table.table_name}", dbConnection)
    df.to_csv(table.version(new_version),index=False)
    print(df.info)  
dbConnection.close()
```
